## 3.1 Exploratory Data Analysis (EDA)

### task:
Performing initial Exploratory Data Analysis by looking at the first few rows, checking for null values, and using `df.describe()` and `df.info()`.


This is descriptive statistics and information about the features DataFrame X to get a better understanding of the data distribution and types.

```python
print("Descriptive statistics of X:")
display(X.describe())

print("\nInformation about X:")
X.info()
```

```python
numerical_cols = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']

for col in numerical_cols:
  Q1 = X[col].quantile(0.25)
  Q3 = X[col].quantile(0.75)
  IQR = Q3 - Q1

  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR

  # Cap outliers
  X.loc[:, col] = X[col].clip(lower=lower_bound, upper=upper_bound)

print("Descriptive statistics of numerical columns after outlier handling:")
display(X[numerical_cols].describe())
```

```python
ca_median = X['ca'].median()
thal_median = X['thal'].median()

X.loc[:, 'ca'] = X['ca'].fillna(ca_median)
X.loc[:, 'thal'] = X['thal'].fillna(thal_median)


print("Missing values in X after imputation:")
print(X.isnull().sum())
```

in this,Converting the target variable to a binary format and then normalizing the features.


Convert the 'num' column in the target DataFrame y to binary values (0 and 1) based on the specified conditions and then display the value counts of the updated 'num' column.

# 3.4 Task
Perform Principal Component Analysis (PCA) on the dataset to reduce dimensionality while retaining 95% of the variance. Normalize the features, implement PCA, plot the explained variance ratio, and output the shape of the reduced dataset.

## Ensure features are normalized

### Subtask:
Verify that the features are normalized, or normalize them if necessary.


**Reasoning**:
Check if X_scaled_df exists and is a DataFrame, and if not, create it from the previously scaled NumPy array X_scaled.



```python
import pandas as pd

# Check if X_scaled_df exists and is a DataFrame
if 'X_scaled_df' not in locals() or not isinstance(X_scaled_df, pd.DataFrame):
    # If not, convert the scaled NumPy array X_scaled back into a DataFrame
    X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

print("Confirmed that the features are in a scaled DataFrame:")
display(X_scaled_df.head())
```

## Implement pca

### Subtask:
Implement pca


**Reasoning**:
Implement PCA by initializing the PCA object and fitting it to the scaled features.



```python
from sklearn.decomposition import PCA

# Initialize PCA without specifying the number of components
pca = PCA()

# Fit PCA to the scaled features
pca.fit(X_scaled_df)

print("PCA fitted to the scaled features.")
```

**Reasoning**:
Plot the explained variance ratio to determine the number of components needed to retain 95% of the variance and output the shape of the reduced dataset.



```python
import numpy as np
import matplotlib.pyplot as plt

# Plot the explained variance ratio
plt.figure(figsize=(10, 6))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance by Number of PCA Components')
plt.grid(True)
plt.show()

# Determine the number of components to retain 95% variance
explained_variance_ratio_cumsum = np.cumsum(pca.explained_variance_ratio_)
n_components_95 = np.argmax(explained_variance_ratio_cumsum >= 0.95) + 1

print(f"\nNumber of components to retain 95% of variance: {n_components_95}")

# Transform the data with the selected number of components
pca_final = PCA(n_components=n_components_95)
X_pca = pca_final.fit_transform(X_scaled_df)

print("\nShape of the reduced dataset after PCA:")
print(X_pca.shape)
```

## Summary:

### Data Analysis Key Findings

*   The features were confirmed to be normalized and stored in a pandas DataFrame.
*   PCA was fitted to the scaled features.
*   A cumulative explained variance plot showed the variance retained as the number of components increased.
*   It was determined that 12 components are needed to retain at least 95% of the variance.
*   The dataset's dimensionality was reduced from its original shape to (303, 12) using PCA with 12 components.



# 3.5 Task
Apply K-means clustering to the PCA-reduced dataset, determine the optimal number of clusters using the Elbow method and Silhouette score, and visualize the clusters on a 2D scatterplot using the first two principal components.

## Apply kmeans clustering

### Subtask:
Apply the KMeans clustering algorithm to the PCA-reduced dataset.


**Reasoning**:
Apply the KMeans clustering algorithm to the PCA-reduced dataset to prepare for determining the optimal number of clusters using the Elbow method.



```python
from sklearn.cluster import KMeans

# Initialize a list to store the inertia values
inertia_values = []

# Loop through a range of possible numbers of clusters
for n_clusters in range(1, 11):
    # Initialize a KMeans object
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10) # Added n_init for reproducible results

    # Fit the KMeans model to the PCA-transformed data
    kmeans.fit(X_pca)

    # Append the inertia_ attribute to the inertia list
    inertia_values.append(kmeans.inertia_)

print("KMeans clustering applied for different numbers of clusters.")
```

## Determine optimal number of clusters (elbow method)

### Subtask:
Use the Elbow method to find a possible optimal number of clusters.


**Reasoning**:
Plot the inertia values against the number of clusters to visualize the elbow method.



```python
import matplotlib.pyplot as plt

# Plot the inertia values against the number of clusters
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia_values, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()

# Based on the plot, the "elbow point" is where the rate of decrease in inertia sharply changes.
# Looking at the plot, there appears to be an elbow around k=3 or k=4.
print("Visual inspection of the plot suggests that a possible optimal number of clusters is around 3 or 4.")
```

## Determine optimal number of clusters (silhouette score)

### Subtask:
Use the Silhouette score to evaluate the quality of clustering for different numbers of clusters.


**Reasoning**:
Calculate and plot the silhouette scores for different numbers of clusters to evaluate clustering quality and identify a potential optimal number of clusters.



```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Initialize an empty list to store the silhouette scores
silhouette_scores = []

# Loop through a range of possible numbers of clusters (from 2 to 10)
for n_clusters in range(2, 11):
    # Initialize a KMeans object
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)

    # Fit the KMeans model to the PCA-transformed data
    kmeans.fit(X_pca)

    # Predict the cluster labels for the data
    cluster_labels = kmeans.predict(X_pca)

    # Calculate the silhouette score
    silhouette_avg = silhouette_score(X_pca, cluster_labels)

    # Append the calculated silhouette score to the silhouette_scores list
    silhouette_scores.append(silhouette_avg)

# Print the calculated silhouette scores for each number of clusters
print("Silhouette scores for different numbers of clusters:")
for i, score in enumerate(silhouette_scores):
    print(f"Number of clusters = {i + 2}: {score:.4f}")

# Plot the silhouette scores against the number of clusters
plt.figure(figsize=(10, 6))
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.title('Silhouette Score for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()
```

## Visualize clusters

### Subtask:
Visualize the clusters on a 2D scatterplot using the first two principal components.


**Reasoning**:
Initialize and fit the KMeans model with the optimal number of clusters, get the cluster labels, and create and display the scatter plot of the first two principal components colored by cluster labels.



```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Initialize a KMeans model with the optimal number of clusters (e.g., 2 based on Silhouette score)
optimal_n_clusters = 2
kmeans = KMeans(n_clusters=optimal_n_clusters, random_state=42, n_init=10)

# Fit the KMeans model to the PCA-transformed data
kmeans.fit(X_pca)

# Get the cluster labels
cluster_labels = kmeans.labels_

# Create a scatter plot of the first two principal components
plt.figure(figsize=(10, 7))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=cluster_labels, cmap='viridis', s=50, alpha=0.6)

# Add labels and title
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('K-means Clustering Visualization on PCA-Reduced Data')

# Add a legend
legend = plt.legend(*scatter.legend_elements(), title="Clusters")
plt.gca().add_artist(legend)


# Display the plot
plt.grid(True)
plt.show()
```

### Data Analysis Key Findings

*   K-means clustering was applied to the PCA-reduced dataset for a range of cluster numbers (1 to 10), and the inertia values were calculated and stored.
*   The Elbow method plot of inertia values against the number of clusters suggested a possible optimal number of clusters around 3 or 4, indicated by a change in the rate of decrease of inertia.
*   The Silhouette score analysis was performed for cluster numbers from 2 to 10. The highest Silhouette score of 0.1775 was achieved with 2 clusters.
*   A 2D scatterplot visualizing the clusters based on the first two principal components was successfully generated using 2 clusters. The data points were colored according to their assigned cluster labels.
