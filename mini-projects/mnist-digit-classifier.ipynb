#NN
using the normal two layer Neural netwrok, analysing the accuracy of model and plotting the graph

```python
# Importing necessary libraries
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
import numpy as np
import matplotlib.pyplot as plt

# Loading the MNIST dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Normalize the image data to [0, 1]
x_train = x_train / 255.0
x_test = x_test / 255.0

# Flatten the images: 28x28 -> 784

# Define the 2-layer Neural Network model
model_nn = Sequential([
    Flatten(input_shape=(28, 28)),     # Input layer that flattens image
    Dense(128, activation='relu'),     # First (hidden) dense layer
    Dense(10, activation='softmax')    # Output layer with softmax activation
])

# Compile the model
model_nn.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',  # Use this as labels are integers (0–9)
    metrics=['accuracy']
)

# Training the model
history = model_nn.fit(x_train, y_train, epochs=5, validation_split=0.1)

# Evaluate on test data
test_loss, test_acc = model_nn.evaluate(x_test, y_test)
print(f"\nTest Accuracy: {test_acc:.4f}")
```

Plot of accuracy vs number of epochs and loss over epochs

```python
plt.figure(figsize=(12, 5))

# Accuracy plot
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.title('Accuracy Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Loss plot
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```

#CNN
i have train model with 32 filter in 1st layer and 64 in 2nd layer and also used maxpoolig to reduce the size of feature map that eventually increased the train and test speed. overall i got 98.97% accuracy.

```python
# Step 1: Import required libraries
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
import matplotlib.pyplot as plt

# Step 2: Load the dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Step 3: Preprocess the data
# Normalize to [0,1]
x_train = x_train / 255.0
x_test = x_test / 255.0

# Add channel dimension (28x28 -> 28x28x1) for Conv2D
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)

# Step 4: Define the CNN model
model = Sequential([
    Conv2D(32, kernel_size=(3,3), activation='relu', input_shape=(28,28,1)),# 1st layer of 32 filter
    MaxPooling2D(pool_size=(2,2)),# maxpooling take max value in (2,2)reduce the size of feature map.

    Conv2D(64, kernel_size=(3,3), activation='relu'),
    MaxPooling2D(pool_size=(2,2)),

    Flatten(),# converts 3d tensor in to 1d
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

# Step 5: Compile the model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Step 6: Train the model
history = model.fit(x_train, y_train, epochs=5, validation_split=0.1)

# Step 7: Evaluate the model
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f"\nTest Accuracy: {test_acc:.4f}")
```

I have plotted the training and test accuracy over epochs. From the graph, we can see that increasing the number of epochs improves the test accuracy, but it also increases the training and test time.

```python
epochs_range = range(1, len(history.history['accuracy']) + 1)

plt.plot(epochs_range, history.history['accuracy'], label='Training Accuracy')
plt.plot(epochs_range, history.history['val_accuracy'], label='Test Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Training vs Test Accuracy over Epochs')
plt.legend()
plt.grid(True)
plt.show()
```

#RNN

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, LSTM, Dense, Dropout
from tensorflow.keras.utils import to_categorical

# Step 1: Load and preprocess data
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Normalize pixel values
x_train = x_train / 255.0
x_test = x_test / 255.0

# Reshape for RNN input (timesteps=28, features=28)
x_train_rnn = x_train.reshape(-1, 28, 28)
x_test_rnn = x_test.reshape(-1, 28, 28)

# One-hot encode the labels
y_train_oh = to_categorical(y_train, 10)
y_test_oh = to_categorical(y_test, 10)

# Step 2: Define LSTM-based RNN model
rnn_model = Sequential([
    LSTM(128, input_shape=(28, 28), return_sequences=False),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(10, activation='softmax')
])

# Step 3: Compile the model
rnn_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Step 4: Train the model
rnn_history = rnn_model.fit(
    x_train_rnn, y_train_oh,
    epochs=10,
    batch_size=128,
    validation_data=(x_test_rnn, y_test_oh)
)

# Step 5: Evaluate the model
test_loss, test_acc = rnn_model.evaluate(x_test_rnn, y_test_oh)
print(f"\nTest Accuracy: {test_acc:.4f}")
```
